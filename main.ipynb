```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("./Earthquakes.csv")
df
```

```python
print(f"first 6 rows\n {df.head()}")
print(f"\n\n COLUMNS \n {df.columns}")
print(f"\n\n VALUES \n {df.values}")
```

```python
sns.pairplot(df, hue='net')
plt.show()
```

```python
data = df.values
x = data[:, 0:4]
y = data[:, 4]
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
sn = SVC()
sn.fit(x_train, y_train)

predict = sn.predict(x_test)
print(predict)
```

```python
model_acc = accuracy_score(y_test, predict)
print(model_acc)
```

```python
x_new = np.array([[-60.8391, -19.9569, 10, 6.4]])
output = sn.predict(x_new)
print(f"the output is {output}")
```
